```python
import numpy as np
import matplotlib.pyplot as plt
import logging

%matplotlib widget

logging.basicConfig(level=logging.INFO)

def polynomial_interpolation(x, y, order, plot=True):
    """
    Interpolates x and y points using a polynomial of the specified order.

    Parameters:
    x (list or numpy array): X points for interpolation
    y (list or numpy array): Y points for interpolation
    order (int): Order of the polynomial for interpolation
    plot (bool, optional): Flag to plot the data using matplotlib (default: False)

    Returns:
    numpy array: Coefficients of the interpolating polynomial
    """

    # Check if the order of the polynomial is valid
    if order >= len(x):
        logging.error("Polynomial order must be less than the number of data points.")
        return None

    # Perform polynomial interpolation
    logging.info("Performing polynomial interpolation of order %d.", order)
    coeffs = np.polyfit(x, y, order)

    # If the plot flag is set, plot the data and interpolation
    if plot:
        plt.scatter(x, y, marker='*', s = 3, color = 'r', label="Data points")
        x_plot = np.linspace(min(x), max(x), 1000)
        y_plot = np.polyval(coeffs, x_plot)
        plt.plot(x_plot, y_plot, linewidth=10, alpha = 0.3, color='b', label=f"Interpolation (Order {order})")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.legend()
        plt.title("Polynomial Interpolation")
        plt.show()

    return coeffs

# Example usage
x = np.linspace(-50,50,5000)


y = x**3

rauschen = np.random.randn(len(y))
y = y + 3*rauschen/100*y
order = 3

coeffs = polynomial_interpolation(x, y, order)
print("Coefficients of the interpolating polynomial:", coeffs)
```
