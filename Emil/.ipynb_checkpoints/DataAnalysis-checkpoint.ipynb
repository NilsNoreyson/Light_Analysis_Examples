# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
%matplotlib widget
```

# Funktionen zum Laden der Daten

```python
def load_data(path):
    # Convert the data section to a pandas DataFrame
    data = pd.read_csv(path, sep=';', header=None, names=['nm', 'I'], skiprows=48)
    data = data[0:-1]
    data.index = data['nm'].astype('float', errors='ignore')
    del data['nm']
    return data
```

# Dataen darstellen und Trainingsdaten erstellen

```python
data_path = Path('Data/')
files = list(data_path.glob('*.csv'))

fig, axe = plt.subplots()

colors = {'blatt':'g', 'whitepaper':'r'}
labels = {'blatt':0, 'whitepaper':1}

label2word = {v:w for w,v in labels.items()}

# Variablen fr Trainingsdaten
X = []
Y = []

for file in files:
    
    sample_type = file.name.split('_')[0]
    
    if sample_type in labels.keys():
        print(file)
        color = colors[sample_type]
        label = labels[sample_type]
        data = load_data(file)
        I = data['I']
        I = I

        X.append(list(I))
        Y.append(label)
        axe.plot(I, color = color, label = file.name)
        
    else:
        data = load_data(file)
        I = data['I'].iloc[::100]
        axe.plot(I, 'k--', label = 'Unknown')
axe.legend()
```

# Training des Neuonalen Netzes

```python
from sklearn.neural_network import MLPClassifier, MLPRegressor
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 3), random_state=1)


reg = MLPRegressor(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 3), random_state=1)


reg.fit(X,Y)
clf.fit(X, Y)
```

```python


files = list(data_path.glob('test_*.csv'))

for f in files:
    data = load_data(f)
    I = list(data['I'])
    prediction = clf.predict([I])

    
    print([label2word[p] for p in prediction][0])
    print(f)
    print(round(reg.predict([I])[0],4))
    print()
```
